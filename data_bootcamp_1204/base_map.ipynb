```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
 
# make sure the value of resolution is a lowercase L,
#  for 'low', not a numeral 1

plt.figure(figsize=(8,8))

my_map = Basemap(projection='ortho', lat_0=50, lon_0=-100,
              resolution='l', area_thresh=1000.0)
 
my_map.drawcoastlines()
my_map.drawcountries()
#my_map.bluemarble()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
 
plt.show()
```

```python
plt.figure(figsize=(8,8))

my_map = Basemap(projection='merc', lat_0=57, lon_0=-135,
    resolution = 'l', area_thresh = 1000.0,
    llcrnrlon=28, llcrnrlat=-12,
    urcrnrlon=42, urcrnrlat=-0)

#my_map.bluemarble()
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='grey',alpha = 0.25,lake_color='aqua')
#my_map.bluemarble()
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))


lon = [39.20 , 40.1760, 35.7516]

lat = [-6.79 , -10.3112, -6.1630]


my_map.drawgreatcircle(lon[0], lat[0], lon[1], lat[1])  

population = [4.365, 0.080, 0.40]

x,y = my_map(lon, lat) #This generated the map coordinates already, so below say fals

#my_map.scatter(x,y, latlon = False, alpha = 0.99, color = "red",marker='D')

labels = ['Dar Es Salam', 'Mtwara', 'Dodoma']
min_marker_size = 20.5

for lon, lat, mag, label in zip(lon, lat, population, labels):
    x,y = my_map(lon, lat)
    msize = mag * min_marker_size
    my_map.plot(x, y, 'ro', markersize=msize, alpha = 0.5)
    plt.text(x+5000, y+5000, label)

    
  
plt.title("Tanzania\n")
 
plt.show()
```

```python
test= pd.merge(demo_tanz, geo_var, how='inner',
                         left_on = "y2_hhid", right_on = "y2_hhid", indicator = True)

# Note by taking the inner, there will be some stuff that is going to be droped. 
# There are no election results for parts of Alaska...

test.head(10)

test.shape
```

```python
plt.figure(figsize=(12,12))

my_map = Basemap(projection='merc', lat_0=57, lon_0=-135,
    resolution = 'l', area_thresh = 1000.0,
    llcrnrlon=28, llcrnrlat=-12,
    urcrnrlon=42, urcrnrlat=-0)

#my_map.bluemarble()
my_map.drawcoastlines()
my_map.drawcountries(linewidth=1.5)
my_map.fillcontinents(color='grey',alpha = 0.25,lake_color='aqua')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))

def get_marker_color(urban):
    if urban == "Urban":
        return ('ro')
    else:
        return ('bo')

min_marker_size = 6.0

for lon, lat, urban in zip(test.lon_modified.tolist(), test.lat_modified.tolist(), test.urban.tolist()):
    x,y = my_map(lon, lat)
    msize = min_marker_size
    mark_color = get_marker_color(urban)
    my_map.plot(x, y, mark_color, markersize=msize, alpha = 0.05)
    
    
plt.title("Tanzania\n")
 
plt.show()

plt.show()
```
